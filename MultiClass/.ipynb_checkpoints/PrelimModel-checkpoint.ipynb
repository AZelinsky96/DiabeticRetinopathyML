```python
import matplotlib.pyplot as plt
import cv2


from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential, load_model
from keras.layers import Activation, Dropout, Flatten, Conv2D, MaxPooling2D, Dense, BatchNormalization
from keras import optimizers 

import warnings 
#warnings.filterwarnings('ignore')

import os
import random 
import numpy as np
from sklearn.metrics import confusion_matrix
import numpy as np
import pandas as pd


np.random.seed(42)
```

```python
## Pulling in some images from the different categories randomly and assessing them. 

def image_display(count = 5, display_img = True, base_path = None, append_img = False): 
    os.chdir(base_path)
    categories = ["Mild", "Moderate", "Severe" , "No_Diabetic_Retinopathy", "Proliferative_Diabetic_Retinopathy"]

    countdown = count
    
    image_sizes = {}
    image_coll  = {}
    for i in categories: 
        if display_img == True:
            print(i)
        image_sizes[i] = []
        image_coll[i] = []
        os.chdir(i)
        while True: 
            rand_n = random.randint(0,len(os.listdir()) -1 )    
            countdown -= 1
            ## Reading in the image
            image = cv2.imread(os.listdir()[rand_n])
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)    
            if display_img == True: 
                plt.imshow(image)
                plt.show()
            image_sizes[i].append(image.shape)
            if append_img == True: 
                image_coll[i].append(image)
            if countdown == 0: 
                countdown = count
                break
    
        os.chdir(base_path)
    os.chdir(base)
    return image_sizes, image_coll
```

```python
def conf_matrix(model, model_type, data_generator): 
    from sklearn.metrics import confusion_matrix
    import numpy as np
    import pandas as pd
    predictions = model.predict_generator(data_generator, steps = np.ceil(data_generator.samples / batch_size))
    
    if model_type == "binary": 
        pred_classes = np.round(predictions)
    else: 
        pred_classes = np.argmax(predictions, axis = 1)
    true_classes = data_generator.classes
    class_labels = list(data_generator.class_indices.keys())  
    conf_matrix = pd.DataFrame(confusion_matrix(true_classes, pred_classes), columns = class_labels)
    #conf_matrix.index = class_labels

    return conf_matrix
```

```python
def plot_results(results_1, accuracy_type = "binary_accuracy", model_name = "NewModel"): 
    import matplotlib.pyplot as plt
    if accuracy_type == "binary_accuracy": 
        validation_acc = "val_binary_accuracy"
        
    epoch = epoch = list(map(lambda x: str(x + 1), list(range(len(results_1.history['loss'])) )))
    fig = plt.figure(figsize = (15,15))
    ax1 = plt.subplot(211)
    ax2 = plt.subplot(212) 
    ax1.plot(epoch, results_1.history[accuracy_type], label = "Training Accuracy", color = 'green')
    ax1.plot(epoch, results_1.history[validation_acc], label = 'Validation Accuracy', color = 'blue')
    ax1.set_title("Comaprison of Training Accuracy\n to Validation Accuracy with Display of Loss")   
    ax1.legend()
    ax1.set_xticklabels([])
    ax1.set_ylabel("Accuracy")
    ax2.plot(epoch, results_1.history['loss'], label = 'Training Loss', color = "green")
    ax2.plot(epoch, results_1.history['val_loss'], label = 'Validation Loss')
    ax2.set_ylabel("Loss Value")
    ax2.set_xlabel("Epoch")
    ax2.set_xticklabels([])
    ax2.legend()
    
    plt.savefig(model_name+".png")
    plt.show()
```

```python
def train_model(model, epochs, train_generator, batch_size, validation_generator, validation_steps): 
    
    import warnings 
    warnings.filterwarnings('ignore')
    results = model.fit_generator(train_generator, 
                              epochs = epochs, 
                              steps_per_epoch = round(train_generator.samples // batch_size), 
                              validation_data = validation_generator,
                              validation_steps = validation_steps)
    return results
```

```python
def display_distributions(classes, directory): 
    os.chdir(directory)
    count = {}
    for p in classes:
        os.chdir(p)
        count[p] = len(os.listdir())
        os.chdir("..")
        
    plt.bar(list(count.keys()), list(count.values()))
    plt.xticks(rotation = "90")
    plt.title("{} Distribution".format(directory.split("/")[-1]))
    plt.show()
    os.chdir(base)
```

```python
def color_gauss(image):
    #import cv2
    sigmaX = 12
    image=cv2.addWeighted( image,4, cv2.GaussianBlur( image , (0,0) , sigmaX) ,-4 ,128)
    return image

```

# Building the Image generator to handle the images

```python
input_shape = (256,256,3)
batch_size = 20

image_gen = ImageDataGenerator(  
                               rotation_range = 10, 
                              width_shift_range= 0.1, 
                              height_shift_range= 0.1, 
                              rescale = 1/255, 
                              shear_range = 0.1, 
                              #zoom_range = 0.05, 
                              horizontal_flip = True, 
                              vertical_flip= True, 
                              fill_mode = "nearest", 
                             preprocessing_function = color_gauss
                              )
image_gen_val = ImageDataGenerator(rescale = 1/255, preprocessing_function = color_gauss)
                               
train_image_gen  = image_gen.flow_from_directory(train_base, 
                                                target_size = input_shape[:2], 
                                                batch_size = batch_size, 
                                                class_mode = "categorical")


test_image_gen   = image_gen_val.flow_from_directory(test_base,
                                                target_size = input_shape[:2], 
                                                batch_size = batch_size, 
                                                class_mode = "categorical"
                                                )
```

## Preliminary Build of Model: 


```python
model = Sequential()



model.add(Conv2D(filters = 32, kernel_size = (5,5),input_shape = input_shape, activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(BatchNormalization())

model.add(Conv2D(filters = 64, kernel_size = (3,3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(BatchNormalization())

model.add(Conv2D(filters = 64, kernel_size = (3,3), activation = "relu"))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(BatchNormalization())

model.add(Conv2D(filters = 96, kernel_size = (3,3), activation = "relu"))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(BatchNormalization())

#model.add(Conv2D(filters = 128, kernel_size = (3,3), activation = 'relu'))
#model.add(MaxPooling2D(pool_size = (2,2)))

"""
model.add(Conv2D(filters = 128, kernel_size = (3,3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2,2)))
"""
model.add(Flatten())


model.add(Dense(512, activation = "relu"))
## Dropout layer to help with overfitting
model.add(Dropout(0.5))


#model.add(Dense(96, activation = 'relu'))
#model.add(Dropout(0.25))
#model.add(Dense(32, activation = 'relu'))
#model.add(Dropout(0.25))

model.add(Dense(5, activation = 'softmax'))




optimizer = optimizers.Adam(lr = 0.001)
model.compile(loss = 'categorical_crossentropy', 
              optimizer = optimizer, 
              metrics = ['accuracy']
             )
```
